```python
import pandas as pd
import numpy as np
```

```python
#read in csv
df= pd.read_csv("amazon_reviews.csv")
```

```python
df.head(5)
```

```python
#change review_time to type datetime
df['reviewTime']= pd.to_datetime(df['reviewTime'])
```

```python
df.info()
```

```python
#take the month and create new column
df['month']= df['reviewTime'].dt.month
df
```

```python
#group the categories by month and get a count of each review per category per month
groupMonth= df.groupby(['category', 'month']).size().reset_index(name= 'count')
groupMonth
```

```python
#create a new dataframe pivoting the columns to rows and rows to columns
groupMonth= groupMonth.pivot(index= 'category', columns= 'month', values= 'count').fillna(0)
groupMonth.head(1)
```

```python
months= range(1, 13) #1 -> 12
month_names= ['January', 'February', 'March', 'April', 'May', 'June', 
              'July', 'August', 'September', 'October', 'November', 'December']

groupMonth= groupMonth.reindex(columns= months, fill_value= 0) #reindexs in case a column doesnt exist
groupMonth.columns= month_names
groupMonth.head(1)
```

```python
groupMonth= groupMonth.reset_index()
groupMonth = groupMonth.astype({month: int for month in month_names})
```
